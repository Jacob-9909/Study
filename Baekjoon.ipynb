```python
# 1002 번 터렛
import math

def intersection_points(x1, y1, r1, x2, y2, r2):
    distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

    # 동심원
    if distance == 0 and r1 == r2:
        return -1
    # 외접 or 내접
    elif distance == abs(r1 - r2) or distance == r1 + r2:
        return 1
    # 교차
    elif abs(r1 - r2) < distance < r1 + r2:
        return 2
    # 그 외
    else:
        return 0
    
T = int(input())

for _ in range(T):
    x1, y1, r1, x2, y2, r2 = map(int, input().split())
    result = intersection_points(x1, y1, r1, x2, y2, r2)
    print(result)
```

```python
# 2869 번 달팽이는 올라가고 싶다

T = input().split(' ')
go = int(T[0])
fall = int(T[1])
night = int(T[2])

if go * (night // (go-fall)) - fall * (night // (go-fall)) >= night:
    time = night // (go-fall) - fall
else:
    time = night // (go-fall) + 1

print(time)
```
